```python
import requests
from bs4 import BeautifulSoup
meses = ["january", "february", "march", "april", "may", "june", "july", "august"]
for mes in meses:


    url = f"https://www.boxofficemojo.com/month/{mes}/2023/?grossesOption=calendarGrosses"
    html = requests.get(url)
    soup = BeautifulSoup(html.text, "html.parser")
    soup
    nombrePelicula = []
    pelicula_items = soup.find_all('td', class_="mojo-field-type-release")

    for item in pelicula_items:
        nombrePelicula.append(item.a.text)

    print (nombrePelicula)
    print (len(nombrePelicula))


    posicion = []
    posicion_items = soup.find_all('td', class_="mojo-field-type-rank")

    for item in posicion_items:
        posicion.append(item.text)

    print (posicion)
    print (len(posicion))
    ingreso=[]
    ingreso_bruto=[]
    ingreso_total=[]

    ingresos_items = soup.find_all('td', class_= "mojo-field-type-money")

    for item in ingresos_items:
        ingreso.append(item.text)

    # Eliminar el símbolo '-' de la lista
    data_cleaned = [item for item in ingreso if item != '-']
    # Eliminar el símbolo '$' de los elementos en el arreglo
    data_without_dollars = [item.replace('$', '') for item in data_cleaned]
    data_without_coma = [item.replace(',', '') for item in data_without_dollars]
    # Convertir los elementos a enteros
    data_integers = [int(item) for item in data_without_coma]

    # Dividir en arreglos de ingreso bruto y ingreso total
    ingreso_bruto = [data_integers[i] for i in range(0, len(data_integers), 2)]
    ingreso_total = [data_integers[i] for i in range(1, len(data_integers), 2)]


    distribuidora = []

    studio_items = soup.find_all('td', class_="mojo-field-type-studio")

    for item in studio_items:
        distribuidora.append(item.text.strip())



    fecha_estreno = []

    fecha_items = soup.find_all('td', class_="mojo-field-type-date")

    for item in fecha_items:
        fecha_estreno.append(item.text.strip())


    datos = []

    for i in range(len(distribuidora)):
        datos.append(
            {
                "Posicion": posicion[i],
                "Pelicula": nombrePelicula[i],
                "Ingreso Bruto": ingreso_bruto[i],
                "Ingreso Total": ingreso_bruto[i],
                "Fecha de estreno": fecha_estreno[i],
                "Distribuidora": distribuidora[i],
                "Mes actual": mes
                
            }
        )
        print (i)

    import pandas as pd

    df = pd.DataFrame(datos)
    df
    año ="2023"
    df.to_csv(f"data/Taquilleras-{mes}-{año}.csv", index=False)
```
